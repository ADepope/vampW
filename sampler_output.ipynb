```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the TSV file
file_path = '/Users/alexanders/Desktop/BayesW_out_bayesw_example_1.tsv'  # Update with your file path
df = pd.read_csv(file_path, sep='\t')

# Create a figure with 4 subplots in a 2x2 grid
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Plot mu vs iter
axs[0, 0].plot(df['iter'], df['mu'], label='mu', color='b')
axs[0, 0].set_ylabel('mu')
axs[0, 0].set_title('mu vs iter')
axs[0, 0].legend()

# Plot alpha vs iter
axs[0, 1].plot(df['iter'], df['alpha'], label='alpha', color='g')
axs[0, 1].set_ylabel('alpha')
axs[0, 1].set_title('alpha vs iter')
axs[0, 1].legend()

# Plot h2 vs iter
axs[1, 0].plot(df['iter'], df['h2'], label='h2', color='r')
axs[1, 0].set_ylabel('h2')
axs[1, 0].set_title('h2 vs iter')
axs[1, 0].legend()

# Plot num_markers vs iter
axs[1, 1].plot(df['iter'], df['num_markers'], label='num_markers', color='m')
axs[1, 1].set_xlabel('iter')
axs[1, 1].set_ylabel('num_markers')
axs[1, 1].set_title('num_markers vs iter')
axs[1, 1].legend()

# Adjust layout
plt.tight_layout()
plt.show()
```
